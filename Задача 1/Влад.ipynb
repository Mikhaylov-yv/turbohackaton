```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score

import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras

GT_RAW = "./../data/Задача 1. Датасет 1. Тренировочный неразмеченный.xlsx" # это файл с датасетом, который предоставлен участникам
GT_FILTERED = "./../data/Задача 1. Датасет 1. Тренировочный размеченный.xlsx" # это датасет, из которого эксперты отфильтровали строки
#SUBM_PATH = "./nikita/turbo_pred_1.csv" # это файл участников, датасет из которого они отфильтровали строки

def metric(y_true, y_pred):
    return f1_score(y_true, y_pred)

def prepare_df(path):
    df = pd.read_excel(path)
    df["Параметр"] = pd.to_datetime(df["Параметр"])
    df.set_index("Параметр", inplace=True)
    return df
```

```python
gt_raw = prepare_df(GT_RAW)
gt_filtered = prepare_df(GT_FILTERED)
#subm_df = prepare_df(SUBM_PATH)

y_gt_raw = pd.DataFrame();
gt_raw["label_true"] = 1 # 1, значит строка должна быть удалена
gt_raw["label_pred"] = 1
gt_raw.loc[gt_filtered.index, "label_true"] = 0 # 0, эти строки остаются
#gt_raw.loc[subm_df.index, "label_pred"] = 0

true = gt_raw["label_true"]
pred = gt_raw["label_pred"]

metric(true, pred)
```

```python
y_gt_raw = gt_raw['label_true']

y_gt_raw
```

```python
gt_raw.drop(["label_true", "label_pred"], axis = 1, inplace = True)
```

```python
gt_raw.describe().transpose()[['mean', 'std']]
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(np.array(gt_raw))
```

```python
print(normalizer.mean.numpy())
```

```python
keras_1_model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(units=10),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
```

```python
keras_1_model.compile(
    optimizer= keras.optimizers.Adam(learning_rate=0.1),
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
%%time
history = keras_1_model.fit(
    gt_raw,
    y_gt_raw,
    epochs=10,
    # Suppress logging.
    verbose=0,
    # Calculate validation results on 20% of the training data.
    validation_split = 0.2)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist
```
